```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("pandas :", pd.__version__)
print("numpy :" , np.__version__)
```

```python

np.random.seed(0)
dates = pd.date_range('20240101', periods=10)
data = pd.DataFrame(np.random.randn(10, 4), index=dates, columns=list('ABCD'))

print(data)


data.plot()
plt.title('Random Data Visualization')
plt.xlabel('Date')
plt.ylabel('Values')
plt.show()
```
